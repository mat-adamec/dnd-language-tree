# Explanation

This program generates a language tree for my Dungeons and Dragons world, Eashavar. It uses graphviz to visualize the tree. The results are a little messy, but sufficient for my purposes. The key below succinctly describes how languages evolved over time, including which languages are still extant in its present time.

### KEY TO READ GRAPH
------------------
Bold, square = root

Diamond = language isolate; may have derivatives, but these are typically constructed

Egg = constructed language

Arrow = divergence

Dot = modernization

Diamond = mutually intelligible (transitive - if a parent is intelligible with a daughter, a daughter is intelligible with all siblings)

Box = convergence or mixture, half-intelligible

None = vague relation (magic)

Curve = constructed

Bold = language in standard use (most modern, used divergence, or broadest mutually intelligible node)


```python
from graphviz import Digraph

#Set up graph. This is where most of the configuration happens.
dot = Digraph(comment='Language Tree', format='svg')
dot.attr(newrank='true', clusterrank='local')
dot.attr(rankdir='LR')
dot.attr(ranksep='1', nodesep='0.5')
dot.attr(label='Eashavar Language Tree', fontsize='48', labelloc='t')
```
